```python
from sdv.single_table import CTGANSynthesizer
from sdv.metadata import SingleTableMetadata
import pandas as pd
```

```python
df = pd.read_excel('Student Social Outing and Spending Survey (Responses).xlsx')

metadata = SingleTableMetadata()
metadata.detect_from_dataframe(df)

model = CTGANSynthesizer(metadata, epochs=300)
model.fit(df)

synthetic_df = model.sample(num_rows=300)
synthetic_df
```

```python
df
```

```python
combined = pd.concat([df, synthetic_df], ignore_index=True)
combined.to_excel('augmented_survey.xlsx', index=False)
```

```python
df = pd.read_excel('augmented_survey.xlsx')

# Q10 - spending midpoints
spend_map = {
    'Less than ₹200': 100,
    '₹200 – ₹500': 350,
    '₹501 – ₹1,000': 750,
    '₹1,001 – ₹2,000': 1500,
    'More than ₹2,000': 2500
}
df['Q10. Approximately how much did YOU personally spend on this outing? (₹) (Include food, travel, entry, shopping — everything)'] = df['Q10. Approximately how much did YOU personally spend on this outing? (₹) (Include food, travel, entry, shopping — everything)'].map(spend_map)

# Q2 - allowance midpoints
allowance_map = {
    'Less than ₹3,000': 1500,
    '₹3,000 – ₹6,000': 4500,
    '₹6,001 – ₹10,000': 8000,
    '₹10,001 – ₹15,000': 12500,
    'More than ₹15,000': 17500
}
df['Q2. What is your approximate monthly allowance / pocket money?'] = df['Q2. What is your approximate monthly allowance / pocket money?'].map(allowance_map)

df
```

```python
from sdv.evaluation.single_table import evaluate_quality

quality_report = evaluate_quality(
    real_data=df,
    synthetic_data=synthetic_df,
    metadata=metadata
)
```

```python
from sdv.evaluation.single_table import run_diagnostic

# This checks if the data format and ranges match your original data
diagnostic = run_diagnostic(
    real_data=df,
    synthetic_data=synthetic_df,
    metadata=metadata
)
```
